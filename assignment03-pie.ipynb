```python
# Libraries

# Data frames. 
import pandas as pd

# Plotting.
import matplotlib.pyplot as plt


import seaborn as sns

# Numerical arrays.
import numpy as np

```

```python
# Loding data

df = pd.read_csv("https://drive.google.com/uc?id=1AWPf-pJodJKeHsARQK_RHiNsE8fjPCVK&export=download")
```

```python
# The data is downloaded so we can inspect it
### source: https://drive.google.com/uc?id=1AWPf-pJodJKeHsARQK_RHiNsE8fjPCVK&export=download

df = df.drop(columns=['Index']) 


# Let's have a look.
df
```

```python
# Making pie charts

# Labels

labels = 'User Id', 'First Name', 'Last Name', 'Sex', 'Emal', 'Phone', 'Date of Birth', 'Job Title'

sizes = [20,10,10,10,10,10,20,10]

fig, ax = plt.subplots()
ax.pie(sizes, labels=labels)
```

```python
# Auto-label slices

fig, ax = plt.subplots()
ax.pie(sizes, labels=labels, autopct='%1.1f%%')
```

```python
fig, ax = plt.subplots()
ax.pie(sizes, labels=labels,
       colors=['olivedrab', 'rosybrown', 'gray', 'saddlebrown', 'chocolate', 'tan', 'darkolivegreen', 'g'])
```

```python
# Hatch slices

fig, ax = plt.subplots()
ax.pie(sizes, labels=labels, hatch=['**O', 'oO', 'O.O', '.||.', '**', 'O**', 'o.O','|.|.'])
```

```python
# staff sex

uniqueInstances = df['Sex'].value_counts()
maleCount = uniqueInstances.iloc[0]
femaleCount = uniqueInstances.iloc[1]

print("Male Count: ", maleCount)
print("Female Count: ", femaleCount)
```

```python
# Gender distribution
# Create a pie chart using the counts of Male and Female
labels = ['Male', 'Female']
sizes = [maleCount, femaleCount]
colors = ['lightblue', 'lightcoral']

plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=colors, explode=(0.1, 0), shadow=True)
plt.title('Gender Distribution')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# Show the plot
plt.show()
```
