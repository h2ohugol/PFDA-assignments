```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the data from the CSV file
url = 'https://raw.githubusercontent.com/andrewbeattycourseware/PFDA-courseware/refs/heads/main/assignment/weatherreadings1.csv'
df = pd.read_csv(url)

# Drop rows with missing values
df = df.dropna(how='any', axis=0)

# Convert reportStartDateTime to datetime format
df['reportStartDateTime'] = pd.to_datetime(df['reportStartDateTime'])

# Set the datetime as the index for easier plotting
df.set_index('reportStartDateTime', inplace=True)

# Initialize the Seaborn style
sns.set(style="whitegrid")

# Set figure size
plt.figure(figsize=(12, 6))  # Adjusted to a smaller figure size

# Plot dry bulb temperature (Celsius) using Seaborn
sns.lineplot(x=df.index, y=df['dryBulbTemperature_Celsius'], label='Dry Bulb Temperature (°C)', color='blue', linewidth=2)

# Plot wet bulb temperature (Celsius) using Seaborn
sns.lineplot(x=df.index, y=df['wetBulbTemperature_Celsius'], label='Wet Bulb Temperature (°C)', color='orange', linestyle='--', linewidth=2)

# Annotate selected points to show values (every 100th point)
for i in range(0, len(df), 100):  # Annotate every 100th point instead of 10th
    dry_temp = df['dryBulbTemperature_Celsius'].iloc[i]
    wet_temp = df['wetBulbTemperature_Celsius'].iloc[i]
    time_point = df.index[i]

    # Annotate dry bulb temperature
    plt.text(time_point, dry_temp, f'{dry_temp:.1f}°C', color='blue', fontsize=8, ha='right')

    # Annotate wet bulb temperature
    plt.text(time_point, wet_temp, f'{wet_temp:.1f}°C', color='orange', fontsize=8, ha='left')

# Add labels and title
plt.xlabel('Time', fontsize=12)
plt.ylabel('Temperature (°C)', fontsize=12)
plt.title('Dry and Wet Bulb Temperatures Over Time (Seaborn)', fontsize=14)

# Customize the legend
plt.legend(loc='best', fontsize=10)

# Improve the appearance of x-axis labels (rotation and alignment)
plt.xticks(rotation=45, ha='right')

# Add grid lines for better readability
plt.grid(True)

# Display the plot
plt.tight_layout()
plt.show()
```
