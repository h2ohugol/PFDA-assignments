```python
# Data frames. 
import pandas as pd

# Plotting.
import matplotlib.pyplot as plt

import seaborn as sns

# Numerical arrays.
import numpy as np
```

### Load data

***

#### Load the weather data from URL.



```python
# Load the weather data

df = pd.read_csv("https://raw.githubusercontent.com/andrewbeattycourseware/PFDA-courseware/refs/heads/main/assignment/weatherreadings1.csv")
```

### The data is loaded so we can inspect it

```python
# drops all NaN instances.
# source: https://stackoverflow.com/questions/44548721/remove-row-with-null-value-from-pandas-data-frame
df = df.dropna(how='any',axis=0) 

# Let's have a look.
df
```

```python
# Inspect types.
df.dtypes
```

```python
# Look at the first row
df.iloc[0]
```

```python
# Describe the data set.
df.describe()
```

```python
# Convert reportStartDateTime to date time formaat
df['reportStartDateTime'] = pd.to_datetime(df['reportStartDateTime'])
```

```python
#Set the date time as the index for easier plotting
df.set_index('reportStartDateTime' , inplace=True)
```

```python
# Create a plot of dry bulb and wet bulb temperatures over time
plt.figure(figsize=(10, 6))
```

```python
# Reduce DPI (Dots Per Inch) to generate smaller image
plt.savefig('plot.png', dpi=100)
```

```python
# Plot dry bulb temperture (Celsius) with a label
plt.plot(df.index, df['dryBulbTemperature_Celsius'], label='Dry Bulb Temperature (°C)', color='blue', linestyle='-', linewidth=2)

```

```python
# Plot wet bulb temperature (Celsius) with a label
plt.plot(df.index, df['wetBulbTemperature_Celsius'], label='Wet Bulb Temperature (°C)', color='green', linestyle='--', linewidth=2)
```
